```python
import pandas as pd 
df = pd.read_excel('/content/drive/MyDrive/Equintel_Uma/agreed_dataset_67K.xlsx')
```

*Find number of positive and negative sentiments 

```python
pos_count = 0
neg_count = 0 
for index,row in df.iterrows():
  if row['sentiment'] == 0:
    neg_count += 1 
  else: 
    pos_count += 1
print('Positives:' + str(pos_count)) 
print('Negatives:' + str(neg_count))
```

To build a balanced dataset, we need equal number of positives and negatives. Since we have 9637 negatives, we also choose only 9637 positives. 

```python
title = []
sentiment = []
count_pos = 0

for index,row in df.iterrows(): 
  if row['sentiment'] == 0: 
    title.append(row['title'])
    sentiment.append(row['sentiment'])
  else: 
    count_pos += 1
    if count_pos < 9637: 
      title.append(row['title'])
      sentiment.append(row['sentiment'])

data = {'title': title,' sentiment' : sentiment}
balanced_df= pd.DataFrame(data)
balanced_df.to_excel("/content/drive/MyDrive/Equintel_Uma/balanced_dataset_67K.xlsx")
```

Validation dataset
