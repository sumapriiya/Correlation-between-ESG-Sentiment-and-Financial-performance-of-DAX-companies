**Read in the Excel file - Validation set** 


```python
import pandas as pd 
df = pd.read_excel('/content/drive/MyDrive/Equintel_Uma/esg_sentiment_data_for_uma_67K.xlsx')
```

**Perform Sentiment Analysis using transformers making use of the Distilbert Model** 

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis',model = 'distilbert-base-uncased-finetuned-sst-2-english')

distilbert = []
for str in df['title']: 
  distilbert.append(classifier(str))
df['distilbert']=distilbert

```

**Update the predicted sentiment and score of the sentiment to the Excel sheet**

```python
distilbert_sentiment = []
distilbert_score = []
for row in df['distilbert']:
  if row[0]['label'] == 'POSITIVE':
    distilbert_sentiment.append(1) 
  else:
    distilbert_sentiment.append(0)
  distilbert_score.append(row[0]['score'])
df['distilbert_sentiment'] = distilbert_sentiment
df['distilbert_score'] = distilbert_score
df.to_excel('/content/drive/MyDrive/Equintel_Uma/esg_sentiment_data_for_uma_67K_updated_181221.xlsx')
```

**Check the number of incorrect predictions **

```python
distil_var_count = 0
for index,row in df.iterrows(): 
  if (row['automated_sentiment'] != row['distilbert_sentiment']):
      distil_var_count+=1

print(distil_var_count)
```

**Perform Sentiment Analysis using Transformers making use of the FinBERT Model**

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis",model="ProsusAI/finbert")
finbert = []
for str in df['title']: 
  finbert.append(classifier(str))
  #print(classifier(str))
df['finbert']=finbert
```

**Update the predicted sentiment and score of the sentiment to the Excel sheet**

```python
finbert_sentiment = []
finbert_score = []
for row in df['finbert']:
  if row[0]['label'] == 'positive':
    finbert_sentiment.append(1) 
  elif row[0]['label'] == 'neutral': 
    finbert_sentiment.append(0.5)
  else:
    finbert_sentiment.append(0)
  finbert_score.append(row[0]['score'])
df['finbert_sentiment'] = finbert_sentiment
df['finbert_score'] = finbert_score
df.to_excel('/content/drive/MyDrive/Equintel_Uma/esg_sentiment_data_for_uma_67K_updated_181221.xlsx')
```

** Convert Neutral score of Finbert model to either positive or negative for easier comparison to the other two models**

```python
# Convert Neutral score of Finbert model to either positive or negative for easier comparison to the other two models
finbert_sentiment_new = []

for row in df['finbert']:
  if row[0]['label'] == 'positive':
    finbert_sentiment_new.append(1) 
  elif (row[0]['label'] == 'neutral' and row[0]['score'] > 0.5): 
    finbert_sentiment_new.append(1)    
  elif (row[0]['label'] == 'neutral' and row[0]['score'] < 0.5):  
    finbert_sentiment_new.append(0) 
  else:
    finbert_sentiment_new.append(0)
 
df['finbert_sentiment_new'] = finbert_sentiment_new
df.to_excel('/content/drive/MyDrive/Equintel_Uma/esg_sentiment_data_for_uma_67K_updated_181221.xlsx')
```

```python
finbert_var_count = 0
for index,row in df.iterrows(): 
  if (row['automated_sentiment'] != row['finbert_sentiment_new']):
      finbert_var_count+=1

print(finbert_var_count)
```
